# 1. Data Cleaning with Pandas:

## Loading Pandas Packages

```python
import pandas as pd
```

##Loading Healthcare Dataset

```python
healthcare_data_cleaning = pd.read_csv('https://raw.githubusercontent.com/nimratmann/datasci_2_manipulation/main/data/healthcare%20data%20cleaning.csv')
healthcare_data_cleaning
```

## Identifying missing values in the dataset
### Changing "missing" to "NaN"

```python
healthcare_data_cleaning.replace('missing', pd.np.nan, inplace=True)
healthcare_data_cleaning
```

## Removing duplicate rows
### There are no duplicate columns to remove

```python
duplicate_rows = healthcare_data_cleaning.duplicated()
healthcare_data_cleaning.drop_duplicates(inplace=True)
healthcare_data_cleaning
```

## Cleaning column names
### Removing White Space and Special Characters


```python
import re

def clean_column_names(healthcare_data_cleaning):
    def clean_name(name):
        cleaned_name = re.sub(r'[^a-zA-Z0-9]', '', name)
        return cleaned_name.lower()


    healthcare_data_cleaning.columns = [clean_name(col) for col in healthcare_data_cleaning.columns]
    return healthcare_data_cleaning

healthcare_data_cleaning = clean_column_names(healthcare_data_cleaning)

print(healthcare_data_cleaning)
```
